#For the purpose of image classifaction, we'll use images of cat and dog; and to train our model to identify them.

```python
from bing_image_downloader import  downloader
```

```python
downloader.download("cat",limit=30,output_dir='images',adult_filter_off=False,)
downloader.download("dog",limit=30,output_dir='images',adult_filter_off=False,)
```

# Preprocessing of data...

```python
#Importing relevant libraries
import os
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread #To read an image
from skimage.transform import resize #To downscale an image

target =[]
images =[]
flat_data=[]
DATADIR =  '/content/images'
categories = ['cat','dog']

for category in categories: 
  class_num = categories.index(category) # Label encoding
  path = os.path.join(DATADIR,category) # Using path to use all the images.
  for img in os.listdir(path):
   img_details=imread(os.path.join(path,img))
   img_resized=resize(img_details,(150,150,3)) # resize the image
   flat_data.append(img_resized.flatten())
   images.append(img_resized)
   target.append(class_num)

flat_data = np.array(flat_data)  
target = np.array(target)
images = np.array(images)
```

#Visualization of our Dataset

```python
uniq,count = np.unique(target,return_counts=True)
plt.bar(categories,count)
plt.show()
```

```python
# Splitting data into train test...
from sklearn.model_selection import train_test_split
xtrain , xtest , ytrain , ytest = train_test_split(flat_data,  target, test_size=0.3, random_state=355)
```

#We'll use Support Vector Machines(SVM) as our model
#Using grid GridSearchCV to find the best parameters for our model

```python
from sklearn.model_selection import GridSearchCV
from sklearn import svm
param_grid = [
              {'C':[1,10,100,1000],'gamma': [0.001, 0.01, 0.1, 1], 'kernel':['linear']},
              {'C':[1,10,100,1000],'gamma':[0.001,0.0001],'kernel':['rbf']},
]
svc = svm.SVC(probability=True)
clf = GridSearchCV(svc,param_grid)
clf.fit(xtrain, ytrain)
```

#Predicting the test data

```python
pred_y = clf.predict(xtest)
pred_y
```

#Accuracy of the model

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
accuracy_score(pred_y,ytest)
```

#Generating classification report

```python
print(classification_report(ytest,pred_y))
```
